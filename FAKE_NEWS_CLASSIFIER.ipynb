```python
voc_size = 5000
```

```python
messages=X.copy()
```

```python
messages.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
```

```python
##preprocess

from nltk.stem.porter import PorterStemmer
corpus = []
ps = PorterStemmer()
for i in range (0, len(messages)):
  review = re.sub("[^a-zA-Z]", " ", messages["title"][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words("english")]
  review = " ".join(review)
  corpus.append(review)
```

```python
from nltk.stem.porter import PorterStemmer ##stemming purpose
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus[1]
```

```python
#one hot

onehot_rep = [one_hot(words, voc_size) for words in corpus]
onehot_rep
```

```python
onehot_rep[1]
```

```python
#EMBEDDING
sent_length=20
EMBEDDED_DOCS = pad_sequences(onehot_rep, padding="post", maxlen=sent_length)
print(EMBEDDED_DOCS)
```

```python
from keras.src.models import model
# model

from keras.layers import SimpleRNN, Embedding, Dense


embedding_vector_features=40
model= Sequential()
model.add(Embedding(voc_size, embedding_vector_features, input_length=sent_length))
model.add(SimpleRNN(100))
model.add(Dense(1, activation="sigmoid"))
model.compile(loss="categorical_crossentropy", optimizer="adam")
model.summary()
```

```python
from keras.src.models import model
# model

from keras.layers import SimpleRNN, Embedding, Dense, Bidirectional


embedding_vector_features=40
model= Sequential()
model.add(Embedding(voc_size, embedding_vector_features, input_length=sent_length))
model.add(Bidirectional(LSTM(100)))
model.add(Dense(1, activation="sigmoid"))
model.compile(loss="binary_crossentropy", optimizer="adam") # Changed loss to binary_crossentropy for binary classification
model.summary()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

X_final = np.array(EMBEDDED_DOCS)
Y_final = np.array(Y)

X_train, X_test, Y_train, Y_test = train_test_split(X_final, Y_final, test_size=0.33, random_state=42)

# Fit the model
# Assuming 'model' variable is already defined from previous steps
model.fit(X_train, Y_train, epochs=10, batch_size=64)

# Now print the summary again to see the built model with parameters
model.summary()
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
y_pred=np.where(y_pred > 0.6, 1,0) ##AUC ROC Curve
accuracy_score(Y_test, y_pred)
```

```python
print(classification_report(Y_test, y_pred))
```
